### Text Data

```python
df_schools['City']=df_schools['City'].str.upper()
df_schools.head()
```

```python
df_schools.columns = df_schools.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(',', '')
df_schools.head()
```

```python
crime_cats=pd.Series(df_crime['PRIMARY_TYPE'].unique()).str.cat(sep=', ')
crime_cats
```

### Missing Data

```python
df_crime_groups=df_crime[['PRIMARY_TYPE', 'ARREST', 'ID']].groupby(['PRIMARY_TYPE', 'ARREST']).count()
df_crime_groups.reset_index(inplace=True)
df_pivot=df_crime_groups.pivot(index='PRIMARY_TYPE', columns='ARREST', values='ID')
df_pivot.head()
```

```python
df_pivot[df_pivot[False].notna() & df_pivot[True].notna()].head()
```

```python
df_pivot[df_pivot[False].isna() | df_pivot[True].isna()].head()
```

```python
df_pivot.fillna(0).head()
```
