```python
bins=[0, 15000, 30000, 50000, 70000, np.inf]
labels=['Very low', 'Low', 'Medium', 'High', 'Very high']
df_census['INCOME_GROUP']=pd.cut(df_census['PER_CAPITA_INCOME '], bins=bins, labels=labels)
df_census.head()
```

```python
grouped_by_income=df_census.groupby('INCOME_GROUP')
```

```python
grouped_by_income.describe().head()
```

```python
grouped_by_income['PER_CAPITA_INCOME '].describe().head()
```

```python
grouped_by_income.size()
```

```python
grouped_by_income.mean()
```

```python
# Multiple aggregates
grouped_by_income.agg([np.mean, np.max])
```

```python
# Different aggregates by column
grouped_by_income.agg({'PER_CAPITA_INCOME ':[np.min, np.max], 'HARDSHIP_INDEX':np.mean})
```

```python
# Apply function on each group (with use of aggregate function)
transformed_groups=grouped_by_income.transform(lambda x: (x-x.mean())/x.std())
```

```python
transformed_groups.mean()
```
