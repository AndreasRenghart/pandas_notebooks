# Combining Dataframes

```python
import pandas as pd
import numpy as np
```

```python
df_crime=pd.read_csv('Chicago_Crime_Data-v2.csv')
df_schools=pd.read_csv('Chicago_Public_Schools_-_Progress_Report_Cards__2011-2012-v3.csv')
df_census=pd.read_csv('Census_Data_-_Selected_socioeconomic_indicators_in_Chicago__2008___2012-v2.csv')
```

```python
df_crime_groups=df_crime[['PRIMARY_TYPE', 'COMMUNITY_AREA_NUMBER', 'ARREST', 'ID']].groupby(['PRIMARY_TYPE', 'COMMUNITY_AREA_NUMBER', 'ARREST']).count()
df_crime_groups.reset_index(inplace=True)
df_crime_groups.head()
```

```python
df_pivot=df_crime_groups.pivot(index='PRIMARY_TYPE', columns='ARREST', values='ID')
df_pivot.head()
```

```python
df_pivot.columns.name=''
df_pivot.columns=['No Arrest', 'Arrest']
df_pivot.fillna(0, inplace=True)
df_pivot=df_pivot.astype('int32')
df_pivot.head()
```

```python
df_stacked=df_pivot.stack()
df_stacked
```

```python
df_stacked.unstack().head()
```

```python
df_pivot.mean()
```
